```python
import time,serial,random
import numpy as np

def servoin(a=98,b=94,c=90,d=100,e=93,f=97,delay=1):
    '''waits end of previous act. and send angles to arduino returns acc-array     
    '''
    while ser.read()!=b'>':
        pass
    acc = ser.readline().split(b'\t')
    A=np.array([[float(acc[0])],[float(acc[1])],[float(acc[2])]])
    print(*A)
    data='in'
    for i in a,b,c,d,e,f,delay:
        try: 
            data+=str(hex(i))[2]+str(hex(i))[3]
        except IndexError:
            data+='0'+str(hex(i))[2]
    ser.write(data.encode())
    print(data)
    
def rand():
    '''random angles'''
    servoin(random.randint(0,180),random.randint(0,180),random.randint(0,180),random.randint(0,180),random.randint(0,180),random.randint(0,180))

def bow():
    '''turn forward'''
    servoin(0,180,90,90,93,97,10)
    servoin(90,90,90,90,93,97,1)

def stamp():
    '''walk on place'''
    servoin(125,125,120,135)
    servoin(60,60,55,65)
    
def quatre_step_walk():
    '''two support'''
    servoin(b=135,d=135)
    servoin(125,125,120,135)
    time.sleep(1)
    servoin(a=60,c=55)    
    servoin(60,60,55,65)
    
def interpreter(st,old):
    '''gets input string and last value of servoin
        returns tuple of updated values servo
        if str='start' to initials
        if str='a=12,c=13' returns (12,94,13,100,93,97,1)
        if str='(12,94,13,100,93,97,1)' returns the same
        if str='(12,94,13,100)' returns the same
    '''
    if st=='start':
        old = [98,94,90,100,93,97,1]
        return old
    try:
        q = list(int(i) for  i in st.split(','))
        old = q
        return old
    except ValueError:
        q = old
        tmp = list(st.split(','))
        for i in tmp: 
            q[ord(str(i).split('=')[0])-97]=int(str(i).split('=')[1])
        old = q
        return old
   
if __name__ == "__main__":
    ser = serial.Serial('COM9', 19200, bytesize=8, parity='N', stopbits=1, timeout=2)
    time.sleep(2)
    print("connected to: "+ ser.portstr)
    old_q=[98,94,90,100,93,97,1]
    while True:
        inp=input()
        old_q=interpreter(inp,old_q)
        servoin(*old_q)
    ser.close()
```
